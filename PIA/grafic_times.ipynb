```python
import os 
import matplotlib.pyplot as plt
```

```python
def createTable(L1X,L1Y,L2X,L2Y,name):
    fig, ax = plt.subplots()
    ax.plot(L1X, L1Y, label="Paralell")
    ax.plot(L2X, L2Y, label="Sequential")
    plt.xlabel("Bytes")
    plt.ylabel("Time (Milisegundos)")
    plt.title("%s Parallel vs Sequential"%name)
    plt.legend()
    plt.show()
```

```python
def readtxt():
    path = "./Tiempos"

    #Datos para  S vs P Encoding
    T_Parallel_Enco = []
    S_Parallel_Enco = []
    S_Sequential_Enco = []
    T_Sequential_Enco =[]

    #Datos para S vs P Decoding
    T_Parallel_Deco = []
    S_Parallel_Deco = []
    T_Sequential_Deco =[]
    S_Sequential_Deco = []
  
    with os.scandir(path) as ficheros:
        for fichero in ficheros:
            name= fichero.name.split("-")
            name2 = name[1].split(".")
            if(name[0] == "tiempo_se"):
                file = open("./Tiempos/%s" % fichero.name, 'r')
                s = file.read()
                T_Sequential_Enco.append(float(s))
                S_Sequential_Enco.append(int(name2[0]))
            if(name[0] == "tiempo_sd"):
                file = open("./Tiempos/%s" % fichero.name, 'r')
                s = file.read()
                T_Sequential_Deco.append(float(s))
                S_Sequential_Deco.append(int(name2[0]))
            if(name[0] == "tiempo_pe"):
                file = open("./Tiempos/%s" % fichero.name, 'r')
                s = file.read()
                T_Parallel_Enco.append(float(s))
                S_Parallel_Enco.append(int(name2[0]))
            if(name[0] == "tiempo_pd"):
                file = open("./Tiempos/%s" % fichero.name, 'r')
                s = file.read()
                T_Parallel_Deco.append(float(s))
                S_Parallel_Deco.append(int(name2[0]))

    return T_Parallel_Enco, T_Sequential_Enco, S_Parallel_Enco, S_Sequential_Enco, T_Parallel_Deco, T_Sequential_Deco, S_Parallel_Deco, S_Sequential_Deco
```

```python

TPE, TSE, SPE, SSE, TPD, TSD, SPD, SSD= readtxt() 

print(TPE, TSE, SPE, SSE, TPD, TSD, SPD, SSD)

createTable(SPE,TPE,SSE,TSE,name="Encoding")

createTable(SPE,TPE,SSE,TSE,name="Decoding")
```
